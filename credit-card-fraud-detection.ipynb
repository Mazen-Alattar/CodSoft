```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score 
from sklearn.metrics import roc_curve
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv('/kaggle/input/fraud-detection/fraudTrain.csv')
```

```python
df.drop(columns=['Unnamed: 0','city','state','zip','street','first','last','trans_date_trans_time','cc_num','dob','trans_num'],inplace=True)
df
```

```python
encoder = LabelEncoder()
df["merchant"] = encoder.fit_transform(df["merchant"])
df["category"] = encoder.fit_transform(df["category"])
df["gender"] = encoder.fit_transform(df["gender"])
df["job"] = encoder.fit_transform(df["job"])
```

```python
df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.countplot(x='is_fraud', data=df)
plt.title('Class Distribution')
plt.xlabel('Is Fraud')
plt.ylabel('Count')
plt.show()
```

```python
x = df.drop(columns=["is_fraud"], inplace = False)
y = df["is_fraud"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,shuffle=True,random_state=40)
```

```python
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
y_prob = model.predict_proba(x_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

plt.figure(figsize=(10, 5))
plt.plot(fpr, tpr, color='darkorange', lw=2)
plt.plot([0, 1], [0, 1], color='navy', lw=2)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('RandomForestClassifier\nAccuracy: {:.2f}%'.format(accuracy * 100))
plt.show()
```
